```python

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd


df = pd.read_csv('my_posts.csv')

model = SentenceTransformer('all-MiniLM-L6-v2')

original_claim = "5G towers are spreading COVID-19"

original_embedding = model.encode(original_claim)
post_embeddings = model.encode(df['post_text'].tolist())

similarities = cosine_similarity([original_embedding], post_embeddings)[0]
df['mutation_score'] = 1 - similarities  # 0 = identical, 1 = completely different

df_sorted = df.sort_values(['platform', 'date'])
print(df_sorted[['platform', 'date', 'post_text', 'mutation_score']])

df_sorted.to_csv('mutation_results.csv', index=False)
```
